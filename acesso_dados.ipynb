```python
import requests
import json
from datetime import datetime
```

```python
def baixar_dados(url):
    response = requests.get(url)
    if response.status_code == 200:
        dados = json.loads(response.content)
        return dados
    else:
        print(f"Erro na requisição: {response.status_code}")

def normalizar_data(data, formato="%Y-%m-%d"):
    return datetime.fromtimestamp(data).strftime(formato)
```

```python
chave = "6bdc7799ee41c9b60aca4214a9acf3c79168ede06d2f1f61c9b1c0260bf7d573"
moedas = ['BTC', 'BNB', 'LUNA', 'MATIC', 'SHIB', 'ETH', 'SOL', 'ATOM', 'AVAX', 'FTM', 'DOT', 'AXS', 'MANA', 'LINK', 'ADA', 'DOGE', 'FTT', 'EGLD']
for moeda in moedas:
    url = f"https://min-api.cryptocompare.com/data/v2/histoday?fsym={moeda}&tsym=USD&limit=1000&api_key={chave}"
    dados = baixar_dados(url)
    data_inicio = normalizar_data(dados['Data']['TimeFrom'])
    data_fim = normalizar_data(dados['Data']['TimeTo'])
    with open(f"./outputs/{data_inicio}_{data_fim}_{moeda}.json", "w") as arquivo:
        json.dump(dados, arquivo)
```
